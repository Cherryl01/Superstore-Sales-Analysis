#Data Cleaning

```python
#Remove duplicates

df.drop_duplicates(inplace=True)
```

```python
#Convert order date to datetime

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
#Create Month-Year column

df['Month_Year'] = df['Order Date'].dt.to_period('M')

df.head()
```

#Featuring Engineering

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create Month_Year column
df['Month_Year'] = df['Order Date'].dt.to_period('M')

# Display results
df[['Order Date', 'Month_Year']].head()
```

```python
df.head()
```

```python
# Extract the main categories
main_category = df['Category']


# Display first 5 rows
main_category.head()
```

#Exploratory Data Analysis

```python
#Category with highest sales
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print(category_sales)

# Visualization

category_sales.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

plt.show()
```

Technology category generates the highest sales.

```python
#most profitable category
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print(category_profit)

# Visualization

category_profit.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)

plt.show()
```

Technology category is the most profitable

```python
#Best perfoming region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(region_sales)

# Visualization

region_sales.plot(kind='bar')

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

plt.show()
```

The West region performs best in revenue.

```python
#monthly sales trend
monthly_sales = df.groupby('Month_Year')['Sales'].sum()

monthly_sales.plot(figsize=(12,5))

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

Sales increase significantly during certain months.

```python
#customer segment analysis
segment_sales = df.groupby('Segment')['Sales'].sum()

print(segment_sales)

# Visualization

segment_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title('Sales by Customer Segment')
plt.ylabel('')

plt.show()
```

Consumer segment contributes the largest percentage of sales.

```python
#top 10 products by sale
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

print(top_products)

# Visualization

top_products.plot(kind='barh')

plt.title('Top 10 Products by Sales')
plt.xlabel('Sales')

plt.show()
```

Reveals the products that generate the highest revenue and contribute most significantly to overall sales performance.

```python
# Correlation heatmap

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))

sns.heatmap(numeric_df.corr(), annot=True)

plt.title('Correlation Heatmap')

plt.show()
```

Shows the strength of relationships between numerical variables, indicating that sales and profit have a positive relationship.
